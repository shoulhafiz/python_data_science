```
# now, i want to build custom aggregation
# let say, maximum of z-score
# z = (x - mean) / std

# help me build z score max
# remove abs
def z_score(x):
  zscores = (x - x.mean())/x.std(ddof=0)
  return zscores.max()

z_score(df['Age'])


#def zscore_max(series):
#  zscores=(series - series.mean()) / series.std()
#  return zscores.max()
```

```
# min max scaling
# minmax = (x - min) / (max - min)
# help me build minmax max

def minmax_max(series):
  minmaxs = (series - series.min()) / (series.max() - series.min())
  return minmaxs.max()

df.groupby(['Survived', 'Sex']).agg({'Fare': [zscore_max, minmax_max],
                            'Age': [zscore_max, minmax_max]})


```

```
# herm, problem with min max scaling, max / max = 100%
# i want to know percentage of maximum value based on standard deviation

def zscore_max(series):
  zscores=(series - series.mean()) / series.std()
  return zscores.max()

def zscore_percentage(series):
  zscores=(series - series.mean()) / series.std()
  return zscores.max() / 3.49 * 100

df.groupby(['Survived', 'Sex']).agg({'Fare': [zscore_max, zscore_percentage],
                            'Age': [zscore_max, zscore_percentage]})
```

```
# remember filter by pandas
# help me

survived = df.loc[df['Survived'] == 1]
not_survived = df.loc[df['Survived'] == 0]

plt.figure(figsize = (5, 3)) # 500 x 300 pixels

plt.scatter(survived['Age'], 
            survived['Fare'], label = 'survived')
plt.scatter(not_survived['Age'], 
            not_survived['Fare'], label = 'not survived',
            alpha = 0.5)
plt.xlabel('age')
plt.ylabel('fare')
plt.title('age vs fare', fontsize = 30)
plt.legend()
plt.show()
```
